```python
from __future__ import print_function
import csv
import glob
import os

from PIL import Image

path_to_images = '/home/chenderong/datasets/miniimagenet/images/'

all_images = glob.glob(path_to_images + '*')
```

```python
path_to_processed_images = "/home/chenderong/datasets/datas/miniImagenet/images/"
for i, image_file in enumerate(all_images):
    im = Image.open(image_file)
    im = im.resize((84, 84), resample=Image.LANCZOS)
    image_save_file = path_to_processed_images+image_file[46:]
    im.save(image_save_file)
    if i % 500 == 0:
        print(i)
```

```python
for datatype in ['train', 'val', 'test']:
    os.system('mkdir ' + datatype)

    with open(datatype + '.csv', 'r') as f:
        reader = csv.reader(f, delimiter=',')
        last_label = ''
        for i, row in enumerate(reader):
            if i == 0:  # skip the headers
                continue
            label = row[1]
            image_name = row[0]
            if label != last_label:
                cur_dir = datatype + '/' + label + '/'
                os.system('mkdir ' + cur_dir)
                last_label = label
            os.system('cp images/' + image_name + ' ' + cur_dir)
```
